# Modelos de Regresion lineal

Grupo 1


```python
#Modelo grupo 1

#Datos de entrada

Horasestudiog1 = np.array([Horasestudiog1]).reshape(-1,1)
Calificaciong1=df1.Calificacion

#Crear y entrenar modelo de regresion lineal
modelog1=LinearRegression()
modelog1.fit(Horasestudiog1,Calificaciong1)

#Obtener la ecuación de la línea de regresión (pendiente e intercepto).
pendiente = modelog1.coef_[0]
intercepto = modelog1.intercept_
print("Ecuación de la línea de regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")

print("Valor de la intercepcion:", modelog1.intercept_)
print("Valor de la pendiente:", modelog1.coef_)


```

```python
#Crear el diagrama de dispersion con la linea de regresion
plt.figure(figsize=(6,5))
plt.scatter(Horasestudiog1, Calificaciong1, color="blue", label="Datos")
plt.plot(Horasestudiog1, modelog1.predict(Horasestudiog1),color="red", label="Linea de regresion")

plt.title("Pronostico de Calificacion a partir de las Horas de Estudio")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificacion")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr
#Datos de entrada
Horasestudiog1=df1.HorasEstudio
Calificaciong1=df1.Calificacion

#Calcular el coeficiente de correlacion de Pearson
coeficiente_correlacion, p_valor = pearsonr(Horasestudiog1, Calificaciong1)

print("Coeficiente de correlacion de Pearson:", coeficiente_correlacion)
```
