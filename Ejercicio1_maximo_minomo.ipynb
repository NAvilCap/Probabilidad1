## Ejemplo  3: Histogramas de 2 grupos


# 1)libreria para numero aleatorio


```python
import random
```

```python
fabrica1=[]
for i in range(100):
   numero_aleatorio=round(random.uniform(50,100), 2)
   fabrica1.append(numero_aleatorio)
   print(fabrica1)
```

```python
fabrica2=[]
for i in range(100):
   numero_aleatorio2=round(random.uniform(50,100), 2)
   fabrica2.append(numero_aleatorio2)
   print(fabrica2)
```

```python
#libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Produccion fabrica 2")
```

```python
media=mean(fabrica1)
mediana=median(fabrica1)
moda=mode(fabrica1)

print ("valores de la fabrica1")
print ("Media=",mediana)
print ("mediana=",media)
print ("moda=",moda)
```

```python
media2=mean(fabrica2)
mediana2=median(fabrica2)
moda2=mode(fabrica2)

print ("valores de la fabrica 2")
print ("Media=",mediana2)
print ("mediana=",media2)
print ("moda=",moda2)
```

```python
# generar subplot de renglon 1 por 2 columnas para desplegar 2 histogramas
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show
```

```python
#valor minimo, valor maximo, rango
media=mean(fabrica1)
mediana=median(fabrica1)
moda=mode(fabrica1)

```

```python
plt.hist(fabrica1,bins=10, color="b",edgecolor="black", alpha=0.4)
plt.axvline(media, color="b", linestyle="dashed", linewidth=2, label="media")
plt.axvline(mediana, color="r", linestyle="dashed", linewidth=2, label="mediana")
plt.axvline(moda, color="y", linestyle="dashed", linewidth=2, label="moda")
plt.xlabel("valor")
plt.ylabel("frecuencia")
plt.title("Produccion diaria")
plt.legend()
plt.show()
```

```python
plt.hist(fabrica2,bins=20, color="b",edgecolor="black", alpha=0.4)
plt.axvline(media, color="b", linestyle="dashed", linewidth=2, label="media")
plt.axvline(mediana2, color="r", linestyle="dashed", linewidth=2, label="mediana")
plt.axvline(moda2, color="y", linestyle="dashed", linewidth=2, label="moda")
plt.xlabel("valor")
plt.ylabel("frecuencia")
plt.title("Produccion diaria")
plt.legend()
plt.show()
```

```python
maximo=max(fabrica1)
minimo=min(fabrica1)
rango=maximo-minimo
media=mean(fabrica1)
mediana=median(fabrica1)
moda=mode(fabrica1)
print("Estadistica descriptiva de la fabrica 1\n","media=",media,"\nmediana=",mediana,"\nmoda=",moda,"\nrango=",rango,"\nmaximo=",maximo,"\nminimo",minimo)
```

```python
maximo2=max(fabrica2)
minimo2=min(fabrica2)
rango2=maximo2-minimo2
media2=mean(fabrica2)
mediana2=median(fabrica2)
moda2=mode(fabrica2)
print("Estadistica descriptiva de la fabrica 2\n","media=",media2,"\nmediana=",mediana2,"\nmoda=",moda2,"\nrango=",rango2,"\nmaximo=",maximo2,"\nminimo",minimo2)
```

 ##Dispersion de datos

```python
desviacion_estandar_fabrica1=stdev(fabrica1)
print("La desviacion estandar de la fabrica 1 es =",desviacion_estandar_fabrica1)
```

```python
desviacion_estandar_fabrica2=stdev(fabrica2)
print("La desviacion estandar de la fabrica 2 es =",desviacion_estandar_fabrica2)
```

```python
plt.hist(fabrica1,bins=20, color="b",edgecolor="black", alpha=0.4)
plt.axvline(media, color="b", linestyle="dashed", linewidth=2, label=f"media={media:.2f}")
plt.axvline(mediana, color="r", linestyle="dashed", linewidth=2, label=f"mediana={mediana:.2f}")
plt.axvline(moda, color="y", linestyle="dashed", linewidth=2, label=f"moda={moda:.2f}")

plt.axvline(media-desviacion_estandar_fabrica1, color="g", linestyle="dashed", linewidth=2, label=f"desviacion={desviacion_estandar_fabrica1:.2f}")
plt.axvline(media+desviacion_estandar_fabrica1, color="g", linestyle="dashed", linewidth=2, )

plt.xlabel("valor")
plt.ylabel("frecuencia")
plt.title("Produccion diaria")
plt.legend()
plt.show()
```
